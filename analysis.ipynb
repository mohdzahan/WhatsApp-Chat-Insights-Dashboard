```python
import pandas as pd
from collections import Counter

def most_common_words(selected_user, df):
    # Read the stopwords file and split them into a list
    with open("stopwords.txt", 'r') as f:
        stop_words = f.read().splitlines()

    # Initialize the list to collect words
    words = []

    if selected_user != 'Overall':
        # Filter messages by the selected user
        df = df[df['users'] == selected_user]

    # Filter out group notifications and media messages
    tempo = df[df['users'] != 'group_notification']
    tempo = tempo[tempo['user_message'] != '<Media omitted>\n']

    # Collect words from user messages
    for message in tempo['user_message']:
        for word in message.lower().split():
            if word not in stop_words:
                words.append(word)

    # Create a DataFrame of the 20 most common words
    most_common_df = pd.DataFrame(Counter(words), columns=['Word', 'Frequency'])
    
    return most_common_df

# Example usage
# df = pd.read_csv('your_data_file.csv')
# print(most_common_words('User123', df))
```

```python
def floyd_cycle(nums):
    if not nums:
        return 0

    slow = nums[0]
    fast = nums[0]


    while True:
        slow = nums[slow]  
        fast = nums[nums[fast]]  
        if slow == fast:
            break


    fast = nums[0]
    while fast != slow:
        fast = nums[fast]
        slow = nums[slow]

    return fast

nums = [3, 4, 2, 1, 4]
print(floyd_cycle(nums))  
```
